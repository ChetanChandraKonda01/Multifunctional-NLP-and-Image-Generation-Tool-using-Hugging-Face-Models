```python
import torch
from transformers import BartForConditionalGeneration, BartTokenizer, GPT2LMHeadModel, GPT2Tokenizer
from transformers import pipeline, AutoModelForQuestionAnswering
from diffusers import StableDiffusionPipeline
from PIL import Image
import torch
from transformers import BartForConditionalGeneration, BartTokenizer, GPT2LMHeadModel, GPT2Tokenizer
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer
from diffusers import StableDiffusionPipeline
from transformers import AutoTokenizer, AutoModelForQuestionAnswering
import torch


# Load models
model_name_bart = "facebook/bart-large-cnn"
tokenizer_bart = BartTokenizer.from_pretrained(model_name_bart)
model_bart = BartForConditionalGeneration.from_pretrained(model_name_bart)

# GPT-2 Model for Next Word Prediction
tokenizer_gpt2 = GPT2Tokenizer.from_pretrained('gpt2')
model_gpt2 = GPT2LMHeadModel.from_pretrained('gpt2')

# Sentiment Analysis Pipeline
sentiment_analysis = pipeline("sentiment-analysis")

# Question Answering Model (BERT)
model_name_qa = "bert-large-uncased-whole-word-masking-finetuned-squad"
tokenizer_qa = AutoTokenizer.from_pretrained(model_name_qa)
model_qa = AutoModelForQuestionAnswering.from_pretrained(model_name_qa)

# Image Generation (Stable Diffusion)
model_id = "CompVis/stable-diffusion-v1-4"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16).to("cuda")
```

```python
# Function for Text Summarization
def summarize(text, model, tokenizer, max_length=130, min_length=30, do_sample=False):
    inputs = tokenizer.encode("summarize: " + text, return_tensors="pt", max_length=1024, truncation=True)
    summary_ids = model.generate(inputs, max_length=max_length, min_length=min_length, length_penalty=2.0, num_beams=4, early_stopping=True)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary

# Function for Next Word Prediction (GPT-2)
def predict_next_word(prompt, model, tokenizer, top_k=5):
    inputs = tokenizer(prompt, return_tensors='pt')
    with torch.no_grad():
        outputs = model(**inputs)
    next_token_logits = outputs.logits[:, -1, :]
    top_k_tokens = torch.topk(next_token_logits, top_k).indices[0].tolist()
    predicted_tokens = [tokenizer.decode([token]) for token in top_k_tokens]
    return predicted_tokens

# Function for Sentiment Analysis
def analyze_sentiment(texts, sentiment_analysis):
    results = sentiment_analysis(texts)
    return results

# Function for Question-Answering
def answer_question(question, context, model, tokenizer):
    inputs = tokenizer.encode_plus(question, context, add_special_tokens=True, return_tensors="pt")
    input_ids = inputs["input_ids"].tolist()[0]
    outputs = model(**inputs)
    answer_start_scores = outputs.start_logits
    answer_end_scores = outputs.end_logits
    answer_start = torch.argmax(answer_start_scores)
    answer_end = torch.argmax(answer_end_scores) + 1
    answer = tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(input_ids[answer_start:answer_end]))
    return answer

# Function for Image Generation
def generate_image(prompt, pipe):
    with torch.no_grad():
        image = pipe(prompt).images[0]
    return image
```

```python
from IPython.display import display, HTML, clear_output
import ipywidgets as widgets
from PIL import Image

# Text Summarization Form
def text_summarization_form():
    text_input = widgets.Textarea(
        description='Text to Summarize:',
        layout=widgets.Layout(width='100%', height='150px')
    )
    summarize_button = widgets.Button(description="Summarize Text")

    def on_summarize_button_click(b):
        clear_output()
        display(text_input, summarize_button)
        text = text_input.value
        if text:
            summary = summarize(text, model_bart, tokenizer_bart)
            print("Summary:", summary)
        else:
            print("Please enter text to summarize.")

    summarize_button.on_click(on_summarize_button_click)
    display(text_input, summarize_button)

# Next Word Prediction Form
def next_word_prediction_form():
    prompt_input = widgets.Text(
        description='Prompt:',
        layout=widgets.Layout(width='100%')
    )
    predict_button = widgets.Button(description="Predict Next Word")

    def on_predict_button_click(b):
        clear_output()
        display(prompt_input, predict_button)
        prompt = prompt_input.value
        if prompt:
            predicted_words = predict_next_word(prompt, model_gpt2, tokenizer_gpt2)
            print("Predicted Next Words:", predicted_words)
        else:
            print("Please enter a prompt.")

    predict_button.on_click(on_predict_button_click)
    display(prompt_input, predict_button)

# Sentiment Analysis Form
def sentiment_analysis_form():
    text_input = widgets.Textarea(
        description='Text for Sentiment:',
        layout=widgets.Layout(width='100%', height='150px')
    )
    analyze_button = widgets.Button(description="Analyze Sentiment")

    def on_analyze_button_click(b):
        clear_output()
        display(text_input, analyze_button)
        text = text_input.value
        if text:
            results = analyze_sentiment([text], sentiment_analysis)
            sentiment = results[0]['label']
            score = results[0]['score']
            print(f"Sentiment: {sentiment}\nScore: {score}")
        else:
            print("Please enter text for sentiment analysis.")

    analyze_button.on_click(on_analyze_button_click)
    display(text_input, analyze_button)

# Question Answering Form
def question_answering_form():
    context_input = widgets.Textarea(
        description='Context:',
        layout=widgets.Layout(width='100%', height='150px')
    )
    question_input = widgets.Text(
        description='Question:',
        layout=widgets.Layout(width='100%')
    )
    answer_button = widgets.Button(description="Get Answer")

    def on_answer_button_click(b):
        clear_output()
        display(context_input, question_input, answer_button)
        context = context_input.value
        question = question_input.value
        if context and question:
            answer = answer_question(question, context, model_qa, tokenizer_qa)
            print("Answer:", answer)
        else:
            print("Please provide both context and question.")

    answer_button.on_click(on_answer_button_click)
    display(context_input, question_input, answer_button)

# Image Generation Form
def image_generation_form():
    prompt_input = widgets.Text(
        description='Prompt for Image:',
        layout=widgets.Layout(width='100%')
    )
    generate_button = widgets.Button(description="Generate Image")

    def on_generate_button_click(b):
        clear_output()
        display(prompt_input, generate_button)
        prompt = prompt_input.value
        if prompt:
            image = generate_image(prompt, pipe)
            image_path = "/content/generated_image.png"
            image.save(image_path)
            display(HTML(f'<img src="{image_path}" width="300"/>'))
        else:
            print("Please enter a prompt.")

    generate_button.on_click(on_generate_button_click)
    display(prompt_input, generate_button)

# Show all the forms together
def show_all_forms():
    print("Welcome to the AI Flask App!\n")
    text_summarization_form()
    next_word_prediction_form()
    sentiment_analysis_form()
    question_answering_form()
    image_generation_form()

# Display all forms
show_all_forms()
```
